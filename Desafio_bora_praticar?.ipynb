# Desafios IMÓVEIS

```python
url_imoveis = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
dados_imoveis = pd.read_csv(url_imoveis, sep=';')
dados_imoveis
```

```python
imoveis_comerciais = ['Conjunto Comercial/Sala',
                      'Prédio Inteiro', 'Loja/Salão',
                      'Galpão/Depósito/Armazém',
                      'Casa Comercial', 'Terreno Padrão',
                      'Loja Shopping/ Ct Comercial',
                      'Box/Garagem', 'Chácara',
                      'Loteamento/Condomínio', 'Sítio',
                      'Pousada/Chalé', 'Hotel', 'Indústria']

df_imoveis = dados_imoveis.query('@imoveis_comerciais not in Tipo')
df_imoveis

df_apartamento = df_imoveis.query('Tipo == "Apartamento"')
df_apartamento
```

```python
df_apartamento['Quartos'].mean().round(2)
```

```python
# len(df_apartamento['Bairro'].unique())
df_apartamento['Bairro'].nunique()
```

```python
df_bairros_ap_mais_caros = df_apartamento.groupby('Bairro')[['Valor']].mean().sort_values(
    'Valor', ascending=False).head()
df_bairros_ap_mais_caros
```

```python
df_bairros_ap_mais_caros.plot(kind='barh', figsize=(14, 10), color='red',
                              xlabel='valor_aluguel', ylabel='bairro')
```

```python
df_apartamento = df_apartamento.fillna(0)

registros_a_remover = df_apartamento.query('Valor == 0 | Condominio == 0').index

df_apartamento.drop(registros_a_remover, axis=0, inplace=True)

df_apartamento.drop('Tipo', axis=1, inplace=True)

df_apartamento
```

```python
selecao1 = df_apartamento['Quartos'] == 1
df_apartamento[selecao1]

selecao2 = df_apartamento['Valor'] < 1200
df_apartamento[selecao2]

selecao_final = (selecao1) & (selecao2)
df_apartamento[selecao_final]
df_1 = df_apartamento[selecao_final]

selecao = (df_apartamento['Quartos'] >= 2) & (df_apartamento['Valor'] < 3000) &(df_apartamento[
    'Area'] > 70)
df_apartamento[selecao]
df_2 = df_apartamento[selecao]
```

```python
df_1.to_csv('dados_apartamentos_df1.csv', index=False, sep=';')
df_2.to_csv('dados_apartamentos_df2.csv', index=False, sep=';')
```
