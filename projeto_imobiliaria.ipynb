# Análise exploratória de dados

## Qual o valor médio de aluguel por tipo de imóvel?

```python
dados.head()
```

```python
dados['Valor'].mean()
# calculando a média dos valores da coluna Valor
```

```python
dados.groupby('Tipo').mean(numeric_only=True)
# agrupando os dados pela coluna Tipo e calculando apenas as médias das colunas com valores
# numéricos
```

```python
dados.groupby('Tipo')['Valor'].mean()
# agrupando os dados pela coluna Tipo e calculando as médias por tipo de imóvel apenas da coluna
# valor
```

```python
dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
# transformando a saída anterior em DataFrame e ordenando por Valor, em ordem crescente
```

```python
df_preco_tipo = dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
# atribuindo à variável df_preco_tipo o comando da célula anterior
df_preco_tipo.plot(kind='barh', figsize=(14, 10), color='purple')
# gerando um gráfico de barras com o seu tipo (horizontal), seu tamanho e sua cor
```

## Removendo os imóveis comerciais

```python
dados.Tipo.unique()
# encontrando os valores únicos de cada Tipo de imóvel
```

```python
imoveis_comerciais = ['Conjunto Comercial/Sala',
                      'Prédio Inteiro', 'Loja/Salão',
                      'Galpão/Depósito/Armazém',
                      'Casa Comercial', 'Terreno Padrão',
                      'Loja Shopping/ Ct Comercial',
                      'Box/Garagem', 'Chácara',
                      'Loteamento/Condomínio', 'Sítio',
                      'Pousada/Chalé', 'Hotel', 'Indústria']
# atribuindo à variável imoveis_comerciais os valores de Tipo de imóvel que são apenas comerciais
```

```python
dados.query('@imoveis_comerciais in Tipo')
# selecionando dentro da coluna Tipo apenas os valores presentes na variável imoveis_comerciais
# o @ é necessário para incluir uma variável dentro de uma expressão no query
```

```python
dados.query('@imoveis_comerciais not in Tipo')
# selecionando dentro da coluna Tipo apenas os valores não presentes na variável imoveis_comerciais
```

```python
df = dados.query('@imoveis_comerciais not in Tipo')
# atribuindo à variável df o comando da célula anterior
df.head()
```

```python
df.Tipo.unique()
# encontrando os valores únicos de cada Tipo de imóvel dentro da variável df
```

```python
df_preco_tipo = df.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
# atribuindo à variável df_preco_tipo o comando da última célula de Qual o valor médio de aluguel
# por tipo de imóvel?
df_preco_tipo.plot(kind='barh', figsize=(14, 10), color='purple')
# gerando um gráfico de barras com o seu tipo (horizontal), seu tamanho e sua cor
```

## Qual o percentual de cada tipo de imóvel na nossa base de dados?

```python
df.Tipo.unique()
```

```python
df.Tipo.value_counts(normalize=True)
# encontrando a quantidade de imóveis para cada Tipo em formato percentual
```

```python
df.Tipo.value_counts(normalize=True).to_frame()
# convertendo para o formato DataFrame
df.Tipo.value_counts(normalize=True).to_frame().rename(
    columns={'proportion':'quantidade_percentual'})
# renomeando a coluna de quantidade de proportion para quantidade_percentual
df.Tipo.value_counts(normalize=True).to_frame().rename(
    columns={'proportion':'quantidade_percentual'}).sort_values('quantidade_percentual')
# ordenando os valores com base na coluna quantidade_percentual
```

```python
df_percentual_tipo = df.Tipo.value_counts(normalize=True).to_frame().rename(
    columns={'proportion':'quantidade_percentual'}).sort_values('quantidade_percentual')
# atribuindo à variável df_percentual_tipo o último comando da célula anterior
df_percentual_tipo.plot(kind='bar', figsize=(14, 10), color='green',
                        xlabel='Tipos', ylabel='Percentual')
# gerando um gráfico de barras com o seu tipo (vertical), seu tamanho, sua cor e com os nomes dos
# eixos x e y
```

## Filtros

### **1. Apartamentos que possuem `1 quarto` e `aluguel menor que 1200`**



```python
df['Quartos'] == 1
# encontrando as linhas onde o valor de Quartos é igual a 1
```

```python
selecao1 = df['Quartos'] == 1
# atribuindo à variável selecao1 o comando da célula anterior
df[selecao1]
# mostrando um DataFrame apenas com as linhas onde o valor na coluna Quartos é igual a 1
```

```python
selecao2 = df['Valor'] < 1200
# atribuindo à variável selecao2 a seleção das linhas da coluna Valor onde os valores são menores
# a 1200
df[selecao2]
# mostrando um DataFrame apenas com as linhas onde os valores na coluna Valor é menor que 1200
```

```python
selecao_final = (selecao1) & (selecao2)
# atribuindo à selecao_final a união das ações nas variáveis selecao1 e selecao2
df[selecao_final]
# mostrando um Dataframe apenas com as linhas que atendem aos requisitos das variáveis selecao1 e
# selecao2
```

```python
df_1 = df[selecao_final]
# atribuindo à variável df_1 o DataFrame gerado na célula anterior
```

### **2. `Apartamentos` que possuem pelo menos `2 quartos`, `aluguel menor que 3000` e `area maior que 70`**

```python
selecao = (df['Quartos'] >= 2) & (df['Valor'] < 3000) & (df['Area'] > 70)
# atribuindo à variável selecao o comando para selecionar os valores desejados nas colunas Quartos,
# Valor e Area
df[selecao]
# mostrando um Dataframe apenas com as linhas que atendem aos requisitos da variável selecao
```

# Manipulando os dados

## Criando colunas numéricas

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
dados = pd.read_csv(url, sep=';')
dados.head()
```

```python
dados['Valor_por_mes'] = dados['Valor'] + dados['Condominio']
dados.head()
```

```python
dados['Valor_por_ano'] = dados['IPTU'] + 12 * dados['Valor_por_mes']
dados.head()
```

## Criando colunas categóricas

```python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro']
dados.head()
```

```python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + \
                      dados['Quartos'].astype(str) + ' quarto(s)' + \
                      ' e ' + dados['Vagas'].astype(str) + ' vaga(s) de garagem.'
dados.head()
```

```python
dados['Possui_suite'] = dados['Suites'].apply(lambda x: "Sim" if x > 0 else "Não")
dados.head()
```

```python
dados.to_csv('dados_completos_dev.csv', index=False, sep=';')
```
